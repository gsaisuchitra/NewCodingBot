```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example sales dataset (You would typically read this from a CSV file or a database)
data = {
    'Product_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Product_Name': ['Laptop XYZ', 'Smartphone ABC', 'Casual T-Shirt', 'Kitchen Blender', 'Tablet Pro', 'Wireless Mouse', 'Shoes ABC', 'Camera'],
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Home Goods', 'Electronics', 'Accessories', 'Clothing', 'Electronics'],
    'Price': [500, 300, 20, 50, 350, 25, 70, 400],
    'Units_Sold': [100, 150, 500, 200, 120, 300, 250, 80],
    'Sales': [50000, 45000, 10000, 10000, 42000, 7500, 17500, 32000],
    'Customer_Segment': ['25-40', '25-40', '18-24', '25-40', '25-40', '18-24', '41-60', '25-40'],
    'Return_Rate': [2, 1.5, 3, 5, 2, 1, 2, 1]  # Percentage of products returned
}
```

```python
df.columns#def overview of the sales data
def basic_overview(df):
    print("Sales Data Overview:")
    print(df.describe())
    print("\nTop 5 Products by Sales:")
    print(df[['Product_Name', 'Sales']].sort_values(by='Sales', ascending=False).head(5))
    print("\nTotal Sales:", df['Sales'].sum())
```

```python
# 2. Total Sales by Category
def sales_by_category(df):
    category_sales = df.groupby('Category')['Sales'].sum().reset_index()
    category_sales = category_sales.sort_values(by='Sales', ascending=False)
    print("\nSales by Category:")
    print(category_sales)
```

```python
 # Plotting sales by category
plt.figure(figsize=(8, 6))
sns.barplot(x='Sales', y='Category', data=df, palette='viridis')
plt.title('Sales by Category')
plt.xlabel('Sales ($)')
plt.ylabel('Category')
plt.show()
```

```python
# 4. Top Products by Units Sold
def top_products_by_units(df):
    top_products = df[['Product_Name', 'Units_Sold']].sort_values(by='Units_Sold', ascending=False).head(5)
    print("\nTop 5 Products by Units Sold:")
    print(top_products)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Units_Sold', y='Product_Name', data=df, palette='pastel')
plt.title('Top Products by Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Product Name')
plt.show()
```

```python
sns.pairplot(df)
```

```python
#boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='Category',y='Price',data=df,color='skyblue')
plt.title('top products by units sold')
plt.show();
```

```python
# 3. Sales by Customer Segment
def sales_by_customer_segment(df):
    segment_sales = df.groupby('Customer_Segment')['Sales'].sum().reset_index()
    segment_sales = segment_sales.sort_values(by='Sales', ascending=False)
    print("\nSales by Customer Segment:")
    print(segment_sales)
```

```python
# Plotting sales by customer segment
plt.figure(figsize=(8, 6))
sns.barplot(x='Sales', y='Customer_Segment', data=df, palette='coolwarm')
plt.title('Sales by Customer Segment')
plt.xlabel('Sales ($)')
plt.ylabel('Customer Segment')
plt.show()
```

```python
# 5. Return Rate by Product
def return_rate_analysis(df):
    return_rate = df[['Product_Name', 'Return_Rate']].sort_values(by='Return_Rate', ascending=False)
    print("\nReturn Rate by Product:")
    print(return_rate)
```

```python
 # Plotting return rate by product
plt.figure(figsize=(8, 6))
sns.barplot(x='Return_Rate', y='Product_Name', data=df, palette='Blues')
plt.title('Return Rate by Product')
plt.xlabel('Return Rate (%)')
plt.ylabel('Product Name')
plt.show()
```

```python
# 6. Correlation between Price and Sales
def price_sales_correlation(df):
    correlation = df[['Price', 'Sales']].corr()
    print("\nCorrelation between Price and Sales:")
    print(correlation)
```

```python
# Scatter plot of Price vs Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price', y='Sales', data=df, hue='Category', palette='Set2')
plt.title('Price vs Sales')
plt.xlabel('Price ($)')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# 7. Running All Analyses
def run_analysis(df):
    basic_overview(df)
    sales_by_category(df)
    sales_by_customer_segment(df)
    top_products_by_units(df)
    return_rate_analysis(df)
    price_sales_correlation(df)

# Run the analysis
run_analysis(df)
```
